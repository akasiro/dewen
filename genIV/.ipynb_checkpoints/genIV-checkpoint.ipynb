```python
# 帮刘老师生成工具变量
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_stata('data.dta')
```

```python
df.shape
```

```python
df1 = df[['firstPublish', 'LessonType', 'price', 'introlen', 'Image', 'Tsimilscore']]
```

```python
df1.head()
```

```python
def processtime(value):
    timepart = re.compile('\d{4}-\d{2}-\d{2}')
    result = re.match(timepart, value)
    tmp = result.group()
    tmp = tmp.replace('-', '')
    tmp = int(tmp)
    return tmp
```

```python
df2 = df1.copy()
df2['firstPublish2'] = df2['firstPublish'].apply(processtime)
```

```python
df2.head()
```

```python
timedf = df2['firstPublish2'].drop_duplicates()
```

```python
timedf.head()
```

```python
timedf = timedf.sort_values(inplace=False).reset_index()
timedf['id'] = timedf.index
timedf = timedf[['firstPublish2', 'id']]
```

```python
df3 = df2.merge(timedf, on='firstPublish2')
```

```python
windows = 365
dict_for_pandas = {'id':[], 'introlen_mean30':[], 'price_mean30':[], 'Image_mean30':[], 'Tsimilscore_mean30':[]}
for i in range(2507):
    j = i+1
    tmp_dict = {}
    tmp_dict['id'] = j
    tmp_df = df3[(df3['id']<j) & (df3['id']>(j-windows-1))]
    tmp_dict['introlen_mean30'] = np.mean(tmp_df['introlen'])
    tmp_dict['price_mean30'] = np.mean(tmp_df['price'])
    tmp_dict['Image_mean30'] = np.mean(tmp_df['Image'])
    tmp_dict['Tsimilscore_mean30'] = np.mean(tmp_df['Tsimilscore'])
    for k in dict_for_pandas.keys():
        dict_for_pandas[k].append(tmp_dict.get(k))
cal_result = pd.DataFrame(dict_for_pandas)
```

```python
dict_for_pandas2 = {'id':[],'LessonType':[], 'introlen_mean30_same':[],'price_mean30_same':[], 'Image_mean30_same':[], 'Tsimilscore_mean30_same':[],'introlen_mean30_nosame':[], 'price_mean30_nosame':[], 'Image_mean30_nosame':[], 'Tsimilscore_mean30_nosame':[]}
for r in df3[['id', 'LessonType']].drop_duplicates().itertuples(index=False):
    tmp_dict = {}
    date_id = getattr(r,'id')
    tmp_dict['id'] = date_id
    tmp_dict['LessonType'] = getattr(r, 'LessonType')
#     print(tmp_dict['LessonType'])
    tmp_df1 = df3[(df3['id']<tmp_dict['id']) & (df3['id']>tmp_dict['id']-windows-1) & (df3['LessonType'] == tmp_dict['LessonType'])]
    tmp_df2 = df3[(df3['id']<tmp_dict['id']) & (df3['id']>tmp_dict['id']-windows-1) & (df3['LessonType'] != tmp_dict['LessonType'])]
    tmp_dict['introlen_mean30_same'] = np.mean(tmp_df1['introlen'])
    tmp_dict['price_mean30_same'] = np.mean(tmp_df1['price'])
    tmp_dict['Image_mean30_same'] = np.mean(tmp_df1['Image'])
    tmp_dict['Tsimilscore_mean30_same'] = np.mean(tmp_df1['Tsimilscore'])
    tmp_dict['introlen_mean30_nosame'] = np.mean(tmp_df2['introlen'])
    tmp_dict['price_mean30_nosame'] = np.mean(tmp_df2['price'])
    tmp_dict['Image_mean30_nosame'] = np.mean(tmp_df2['Image'])
    tmp_dict['Tsimilscore_mean30_nosame'] = np.mean(tmp_df2['Tsimilscore'])
    for k in dict_for_pandas2.keys():
        dict_for_pandas2[k].append(tmp_dict.get(k))
cal_result2 = pd.DataFrame(dict_for_pandas2)
```

```python
results = cal_result.merge(cal_result2, on=['id'], how = 'outer')
results = timedf.merge(results, on = 'id')
```

```python
print(results.shape)
results.head()
```

```python
final_results = df2.merge(results, on = ['firstPublish2', 'LessonType'])
```

```python
final_results.shape
```

```python
final_results.to_csv('data_dewen365.csv', index=False)
```
